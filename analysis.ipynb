```python
st.columns = st.columns.str.strip().str.lower()

st['date'] = pd.to_datetime(st['date'], dayfirst=True).dt.date

print(st.head())
```

```python
st['classification'] = st['classification'].replace({
    'Extreme Fear': 'Fear',
    'Extreme Greed': 'Greed'
})
```

```python
st = st[['date', 'classification']]
```

```python
st['date'] = pd.to_datetime(st['date']).dt.date
```

```python
print("Trader:", tr['date'].min(), tr['date'].max())
print("Sentiment:", st['date'].min(), st['date'].max())
```

```python
merged = pd.merge(tr, st, on='date', how='inner')
```

```python
print(merged.shape)
print(merged['classification'].value_counts())
```

```python
merged = merged[merged['classification'] != 'Neutral']
```

```python
merged['classification'].value_counts()
```

```python
merged.groupby('classification')['Closed PnL'].mean()
```
